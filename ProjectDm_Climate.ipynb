# **RMUTP Climate Change Prediction**

---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

**Surface Pressure** ความดันพื้นผิวบริเวณมอ

```python
Surface_df = pd.read_csv('SurfacePS.csv')
Surface_df.head()
```

```python
X = Surface_df['YEAR'].values.reshape(-1, 1)
Y = Surface_df['AVG'].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2024, The predicted mean surface pressure change for Rmutp would be {lr.predict([[2024]])[0][0]}(kPa)')
print('Mean Absolute Error',metrics.mean_absolute_error(Y, Y_pred))

plt.scatter(X, Y)
plt.title('Surface Pressure')
plt.xlabel('Years')
plt.ylabel('Mean Surface Pressure (kPa)')
plt.plot(X, Y_pred, color='red')
plt.show()
```

**Temperature** อุณหภูมิบนพื้นดินบริเวณมอ

```python
Temp_df = pd.read_csv('Temp.csv')
Temp_df.head()
```

```python
X = Temp_df['YEAR'].values.reshape(-1, 1)
Y = Temp_df['AVG'].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2024, The predicted mean surface temperature change for Rmutp would be {lr.predict([[2024]])[0][0]}°C')
print('Mean Absolute Error',metrics.mean_absolute_error(Y, Y_pred))

plt.scatter(X, Y)
plt.title('Temperature')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Years')
plt.plot(X, Y_pred, color='red')
plt.show()
```
